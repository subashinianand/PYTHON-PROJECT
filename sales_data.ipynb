```python
import pandas as pd
df=pd.read_csv(file_path)
df.head()
```

```python
df.info()
```

```python
df.columns
```

```python
sales_array=df["Sales_Amount"].to_numpy()
quantity_array=df["Quantity_Sold"].to_numpy()
df['profit']=df['Sales_Amount'] - (df['Quantity_Sold']*df['Unit_Cost'])
print("sales Array:",sales_array[:5])
```

```python
import numpy as np
total_sales = np.sum(sales_array)
average_sales = np.mean(sales_array)

print("Total Sales:", total_sales)
print("Average Sales:", average_sales)
```

```python
print("First 5 Sales:", sales_array[:5])
```

```python
print("First 5 sales:",sales_array[-5:])
```

```python
mean_sales = np.mean(sales_array)
median_sales = np.median(sales_array)
variance_sales = np.var(sales_array)
std_sales = np.std(sales_array)

print("Mean:", mean_sales)
print("Median:", median_sales)
print("Variance:", variance_sales)
print("Standard Deviation:", std_sales)
```

```python
price_array = df["Unit_Price"].to_numpy()

revenue_array = quantity_array * price_array

print("Revenue (first 5):", revenue_array[:5])
```

```python
profit_array = df['profit'].to_numpy()
profit_margin = (profit_array / sales_array) * 100
print("Profit Margin % (first 5):", profit_margin[:5])
```
